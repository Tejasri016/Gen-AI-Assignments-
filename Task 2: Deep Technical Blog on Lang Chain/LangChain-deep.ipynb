# 🤖 2. Local LLM using HuggingFace

```python
from transformers import pipeline

# Load local model
generator = pipeline("text-generation", model="gpt2")

def llm_response(prompt):
    result = generator(prompt, max_length=100, num_return_sequences=1)
    return result[0]['generated_text']

print(llm_response("Explain LangChain in simple terms"))
```

# 📝 3. Prompt Template

```python
from langchain_core.prompts import PromptTemplate

template = PromptTemplate.from_template(
    "Explain {topic} like I'm a beginner"
)

prompt = template.format(topic="Artificial Intelligence")
print(prompt)
print(llm_response(prompt))
```

# 🔗 4. Chains (Manual Chain Simulation)

```python
def chain_run(topic):
    prompt = template.format(topic=topic)
    return llm_response(prompt)

print(chain_run("Machine Learning"))
```

# 🧠 5. Memory (Simple Simulation)

```python
memory = []

def conversation(input_text):
    memory.append(input_text)
    context = " ".join(memory)
    return llm_response(context)

print(conversation("Hi, my name is Tejasri"))
print(conversation("What is my name?"))
```
